### Optimization in Higher Dimensions

$$f(x, y) = x^2 + y^2$$

```python
def f(x, y):
  return x**2 + y**2

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

X, Y = np.meshgrid(x, y)
# X[0]
```

```python
ax = plt.axes(projection = '3d')
# ax.plot3D(x, y, f(x, y))
ax.plot_surface(X, Y, f(X, Y), alpha = 0.3)
ax.contour3D(X, Y, f(X, Y), zdir = 'x', offset = -2 )
```

```python
X, y = make_blobs(n_samples = 1000,  random_state = 22)
X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.mixture import GaussianMixture
```

```python
GaussianMixture?
```

```python
means = mix.means_
```
