```r
water <- read_csv('water_potability.csv')
```

```r
head(water, 2)
```

```r
model <- glm(Potability ~ ., data = water, family = 'binomial')
```

```r
summary(model)
```

```r
plot(model)
```

```r
head(water, 2)
```

```r
reg_model <- lm(Solids ~ ph + Potability, data = water)
```

```r
plot(reg_model)
```

```r
#model performance???
preds <- predict(model, water)
```

```r
water$Potability
```

```r
ifelse(preds < -.5, 1, 0)
```

#### Chi-Squared test and distribution

```r
1 - pchisq(3, df = 4)
```

```r
expected_absences <- c(12, 12, 12, 12, 12)
observed_absences <- c(15, 12, 9, 9, 15)
```

```r
?table
```

```r
M <- as.table(rbind(observed_absences, expected_absences))
```

```r
chisq.test(M)
```
