```r
data <- read.csv('credit_data.csv')
model <- lm(Balance ~ Income, data = data)
print(summary(model))
```

```r
model <- lm(Balance ~ Income + Student, data = data)
model
```

```r
ggplot(credit, aes(x = Income, y = Balance, color = Student)) +
  geom_point() +
  geom_smooth(method = "lm", se = FALSE)
```

```r
model3 <- lm(Balance ~ Income + Student + Income * Student, data = data)
model3
```

```r
model4 <- lm(Balance ~ Region, data = data)
model4
```

```r
model5 <- lm(Balance ~ Region + Income + Income*Region, data = data)
model5
```
